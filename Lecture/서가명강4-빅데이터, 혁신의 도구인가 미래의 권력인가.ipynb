- 반도체 공정 이상 탐지
- 기계 이상 탐지

- 정상적인 상황을 학습하여, "어디까지가 정상인지" 이해한 후,
- 비정상적인 상활이 발생하였을 때, 반응함


- 원인 vs 예측
    - 예측
        - 100% 예측 불가, 귀납적 방법은 100%가 불가능.
        - 예를 들어, 한 번도 경험해보지 않은 스팸... 새로운 유형이 들어오면, ML은 막을 수 없다.
        - ML은 과거에 일어났던 걸 잘 학습해서 배운 것이다.
        - 게으른 학생, 배운 것만 공부한 학생.
        - 학습하지 않은 건 전혀 모른다.
        - 처음 일어난 일은 모른다.
        - 축구 경기처럼 수많은 팀들이 수많은 경기기록이 있다거나, 수 많은 기사들이 수많은 경기들의 기록이 담긴 바둑은 가능.
        
        
        - 데이터사이언스는 상관관계를 찾는 것
            - 현상에 대한 이해를 가진 사람이 판단하는 것이다.
            - 수많은 상관관계 중에 인과관계가 있을 것인데... 그 중에 ML이 추려서 사람이 고르는 것!
            
            - 치약-칫솔이 많이 팔린 데이터가 있어서, 살펴보면 세트로 묶어서 프로모션을 한 경우도 있다...
            
            
        - 데이터의 한계
            - 퇴사자 예측
                - 21세기 북스 <왜 일>
                    - 퇴사자의 80%가 상사와의 갈등 때문에...
                    - 본질적으로 데이터로 정확한 예측이 어려움.
                    - 확보한 데이터에서 리스크를 20%라도 줄일 수 있으니까!
