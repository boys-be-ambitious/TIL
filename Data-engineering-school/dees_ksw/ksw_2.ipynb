````python
import ipywidgets as widgets
import geopy
import gmaps
import json
from h3 import h3

API_KEY = "-----" # API키는 각자가 발급받아서 사용

gmaps.configure(api_key=API_KEY)

seoul_center = [37.5430402,126.9777753]

class H3Drawer(object):

    def __init__(self):
        self._figure = gmaps.figure(
            center=seoul_center, 
            zoom_level=12, 
            layout={'width': '800px', 'height': '600px'})
        self._container = widgets.Box([self._figure])

        h3_point = h3.geo_to_h3(lat=seoul_center[0], lng=seoul_center[1], res=4)
        childrens = h3.h3_to_children(h3_point, res=5)
        
        seoul_geo_json_file = open('/Users/swkim/Source/seoul-maps/juso/2015/json/seoul_municipalities_geo_simple.json') # 파일 이름을 구글에서 검색하면, github repo 가 있음

        seoul_geo_json = json.load(seoul_geo_json_file)
        
        seoul_gu_list = seoul_geo_json['features']
        print('number of gu: {}'.format(len(seoul_gu_list)))
        for gu_idx in range(len(seoul_gu_list)):
            gu_geometry = seoul_gu_list[gu_idx]['geometry']
            
            gu_kor_name = seoul_gu_list[gu_idx]['properties']['SIG_KOR_NM']
            gu_eng_name = seoul_gu_list[gu_idx]['properties']['SIG_ENG_NM']
                        
            # fixing reversed lat,lng
            coords = gu_geometry['coordinates'][0]
            for i in range(len(coords)):
                arr = coords[i]
                reversed_lat_lng = [arr[1], arr[0]]
                gu_geometry['coordinates'][0][i] = reversed_lat_lng

            polyfilled_points = list(h3.polyfill(geo_json=gu_geometry, res=7))

            print('number of hexagons in {}: {}'.format(gu_kor_name, len(polyfilled_points)))
            
            for h3_point in polyfilled_points:
                boundary = h3.h3_to_geo_boundary(h3_point)
                boundary_tuple = list(map(lambda arr: (arr[0], arr[1]), boundary))

                # draw polygon layer
                drawing = gmaps.drawing_layer(
                    features=[
                        gmaps.Polygon(
                            boundary_tuple,
                            fill_color='red'
                        )
                    ],
                    show_controls=False
                )
                self._figure.add_layer(drawing)
            
    def render(self):
        return self._container

H3Drawer().render()``
````
