    - OOP (Object Oriented)
        - 프로그래밍을 작성할 떄의 방법론

        - 클래스는 blue print
            - object 특성을 기술서
            - instance를 만들어내는 특성
            
    - Dynamic Type
        - Type : 언어의 기초적인 Element들이 있는데 그것들의 종류를 Type이라고 한다.
        - Variable을 미리 선언할 필요가 없음
